# Selection Sort


```python
def Selection_sort(x):
    n = len(x)
    for i in range(n):
        min_value = i                           # here index is considered i.e. i and not x[i]
        for j in range(i+1,n):                  # i put n-i-1 whereas i+1,n was needed 
            if x[j] < x[min_value]:
                min_value = j                   #here index is considered than its actual value
        x[i],x[min_value] = x[min_value],x[i]   #here x[min_value] is needed for swapping and its placement was wrong
    return x                                    # was returning min_value instead of x 
                        
x = [13,46,24,52,20,9]
result = Selection_sort(x)
print(result)
```

# Bubble Sort

```python
def bubble_sort(x):
    for i in range(len(x)):
        for j in range(len(x)-i-1):           ##
            if x[j] > x[j+1]:
                x[j],x[j+1] = x[j+1],x[j]
    return x
    
x = [13,46,24,52,20,9]
result1 = bubble_sort(x)
print(result1)
```

# Insertion sort

```python
def insertion_sort(a):
    for i in range(1, len(a)):
        current = a[i]
        j = i-1                 ###
        while j>=0 and a[j]>current:
            a[j+1] = a[j]
            j -= 1
        a[j+1] = current
    return a


a = [5,9,14,45,15,3]
result = insertion_sort(a)
print(result)
```

# Merge sort


```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]
        
        merge_sort(left)  # Recursively sort left half [3][5]
        merge_sort(right) # Recursively sort right half
        
        # Merge process
        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
            else:
                arr[k] = right[j]
                j += 1
            k += 1
        
        # Handle remaining elements
        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1
        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1
    return arr

arr = [3,1,7,4,5,9,2]
result = merge_sort(arr)
print(arr)
```

# Quick sort


```python
def partition(arr, low, high):
    pivot = arr[high]  # Last element as pivot [1][3]
    i = low - 1  # Index for smaller elements
    
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]  # Swap [1][3]
    
    arr[i+1], arr[high] = arr[high], arr[i+1]  # Place pivot correctly
    return i + 1

def quick_sort(arr, low=0, high=None):
    if high is None:
        high = len(arr) - 1
    if low < high:
        pivot_idx = partition(arr, low, high)
        quick_sort(arr, low, pivot_idx-1)  # Sort left subarray [1][3]
        quick_sort(arr, pivot_idx+1, high)  # Sort right subarray

# Example usage:
arr = [64, 34, 25, 12, 22, 11, 90, 5]
quick_sort(arr)
print("Sorted array:", arr)  # Output: [5, 11, 12, 22, 25, 34, 64, 90]
```
