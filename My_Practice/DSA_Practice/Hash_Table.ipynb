```python
stock_prices = []
with open("stock_prices.csv","r") as f:
    for line in f:
        #print(line)
        tokens = line.split(',')
        # print(tokens[1])
        # print(type(tokens[1]))
        day = tokens[0]
        price = float(tokens[1])
        stock_prices.append([day,price])
        
```

```python
for element in stock_prices:
    #print(element)
    if element[0] == "march 6":
        print(element[1])
```
