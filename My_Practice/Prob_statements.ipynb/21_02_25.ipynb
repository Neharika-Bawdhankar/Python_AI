#Read First n Lines
Create a function that reads and returns the first n lines of a file. Handle cases where n exceeds the file's line count.

```python
with open ("sample.txt",'r+') as file2:
    n = 2
    file2.read(n)
    for line in enumerate(file2):
        print(line)
```

```python
def read_first_n_lines(filename, n=2):
    try:
        with open(filename, 'r') as file:
            lines = []
            for line_num, line in enumerate(file, 1):
                if line_num > n:
                    break
                lines.append(line.strip())
            return lines[:n]  # Ensures correct count if file has <n lines
    except FileNotFoundError:
        return ["Error: File not found"]
    except Exception as e:
        return [f"Error: {str(e)}"]

# Example usage
print(read_first_n_lines("sample.txt", 2))
```

#Given a positive integer n, count the number of digits in n that divide n evenly (i.e., without leaving a remainder). Return the total number of such digits.

A digit d of n divides n evenly if the remainder when n is divided by d is 0 (n % d == 0).
Digits of n should be checked individually. If a digit is 0, it should be ignored because division by 0 is undefined.

```python
def evenlyDivides(n):
    final_list = []
    list1= list(map(int,str(n)))
    for i in list1:
        # if n[i] == 0:
        #     continue
        if i != 0 and n % i == 0:
            final_list.append(i)

    total = len(final_list)
    return total
        
    
n = 23 
result = evenlyDivides(n)
print(result)
```

#Reverse Digits of A Number
Problem Statement: Given an integer N return the reverse of the given number.

Note: If a number has trailing zeros, then its reverse will not include them. For e.g., reverse of 10400 will be 401 instead of 00401.

```python
def reverse_digit(n):
    revNum = 0
    while n > 0:
    # Extract the last digit of
    # 'n' and store it in 'ld'.
        ld = n % 10
        # Multiply the current reverse number
        # by 10 and add the last digit.
        revNum = (revNum * 10) + ld
        # Remove the last digit from 'n'.
        n = n // 10
# Print the reversed number.
    return revNum
        
    
n = 123
result1 = reverse_digit(n)
print(result1)
```

##Print N to 1 using Recursion

```python
def print_numbers_reverse(n):
    if n < 1:
        return
    print(n)
    print_numbers_reverse(n-1)

print_numbers_reverse(5)  # Prints 5-1
```

##Sum of first N Natural Numbers

```python
def sum_numbers(n,i,s):
    if i > n:
        return s 
    s += i
    return sum_numbers(n,i+1,s)

n = 5   
print(sum_numbers(n, i = 1 , s=0))
```

#Factorial of a Number : Iterative and Recursive

```python
def factorial_of_number(n):
    if(n==0):
        return 1
    return n * factorial_of_number(n-1) 
    
n = 5
factorial_of_number(n)
```
