##2. Inheritance Hierarchy
Implement:

Parent class Shape with abstract area() method

Child classes Circle and Rectangle implementing area()

Add perimeter() method to Rectangle only

Target Concept: Inheritance, abstraction

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    """Abstract base class for geometric shapes"""
    @abstractmethod
    def area(self) -> float:
        pass

class Circle(Shape):
    def __init__(self, radius: float):
        self.radius = radius
    
    def area(self) -> float:
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width: float, height: float):
        self.width = width
        self.height = height
    
    def area(self) -> float:
        return self.width * self.height
    
    def perimeter(self) -> float:
        return 2 * (self.width + self.height)

# Example usage
if __name__ == "__main__":
    circle = Circle(5)
    rectangle = Rectangle(4, 6)
    
    print(f"Circle area: {circle.area():.2f}")  # Output: 78.54
    print(f"Rectangle area: {rectangle.area()}")  # Output: 24
    print(f"Rectangle perimeter: {rectangle.perimeter()}")  # Output: 20
```

```python
class InsufficientFundsError(Exception):
    """Custom exception for withdrawal attempts exceeding balance"""
    pass

class BankAccount:
    total_accounts = 0  # Class attribute
    
    def __init__(self, account_number: str, initial_balance: float = 0):
        self.__account_number = account_number  # Private attribute
        self.__balance = initial_balance  # Private attribute
        BankAccount.total_accounts += 1
        
    def deposit(self, amount: float) -> None:
        if amount <= 0:
            raise ValueError("Deposit amount must be positive")
        self.__balance += amount
        
    def withdraw(self, amount: float) -> None:
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive")
        if amount > self.__balance:
            raise InsufficientFundsError(f"Available balance: {self.__balance}, tried to withdraw: {amount}")
        self.__balance -= amount
        
    def get_balance(self) -> float:
        return self.__balance
    
    def get_account_number(self) -> str:
        return self.__account_number
    
    @classmethod
    def get_total_accounts(cls) -> int:
        return cls.total_accounts

# Example usage
if __name__ == "__main__":
    try:
        acc1 = BankAccount("GB1234", 1000)
        acc2 = BankAccount("GB5678", 500)
        
        acc1.deposit(200)
        acc1.withdraw(300)
        print(f"Account {acc1.get_account_number()} balance: £{acc1.get_balance()}")
        
        # This will raise InsufficientFundsError
        acc2.withdraw(600)
        
    except InsufficientFundsError as e:
        print(f"Transaction failed: {e}")
    finally:
        print(f"Total bank accounts: {BankAccount.get_total_accounts()}")
```

#4. Employee Management
Implement:

Employee base class with name, employee_id

Manager subclass with team_size attribute

Developer subclass with tech_stack list

Override display_info() for each subclass

Target Concept: Polymorphism, method overriding

```python
class Employee:
    """Base class for all employees"""
    def __init__(self, name: str, employee_id: str):
        self.name = name
        self.employee_id = employee_id
        
    def display_info(self) -> str:
        return f"ID: {self.employee_id}\nName: {self.name}"

class Manager(Employee):
    def __init__(self, name: str, employee_id: str, team_size: int):
        super().__init__(name, employee_id)
        self.team_size = team_size
        
    def display_info(self) -> str:
        base_info = super().display_info()
        return f"{base_info}\nRole: Manager\nTeam Size: {self.team_size}"

class Developer(Employee):
    def __init__(self, name: str, employee_id: str, tech_stack: list):
        super().__init__(name, employee_id)
        self.tech_stack = tech_stack
        
    def display_info(self) -> str:
        base_info = super().display_info()
        tech_list = ", ".join(self.tech_stack)
        return f"{base_info}\nRole: Developer\nTech Stack: {tech_list}"

# Demonstration of polymorphism
if __name__ == "__main__":
    employees = [
        Manager("Sarah Johnson", "MGR-002", 8),
        Developer("Alex Chen", "DEV-157", ["Python", "AWS", "Docker"]),
        Manager("Bob Wilson", "MGR-011", 12),
        Developer("Priya Gupta", "DEV-198", ["Java", "Spring Boot"])
    ]

    for employee in employees:
        print(employee.display_info())
        print("-" * 30)
```

## 5. Library System
Create:

Book class with title/author/status (available/borrowed)

Library class managing book inventory

Methods: checkout_book(), return_book(), search_by_title()

Use composition (Library contains Books)

Target Concept: Composition, system design

```python
class BookNotFoundError(Exception):
    """Raised when a book operation fails due to missing inventory"""
    pass

class BookNotAvailableError(Exception):
    """Raised when trying to checkout an unavailable book"""
    pass

class Book:
    def __init__(self, title: str, author: str):
        self.title = title
        self.author = author
        self.status = "available"
        
    def __repr__(self):
        return f"«{self.title}» by {self.author} ({self.status})"

class Library:
    def __init__(self):
        self._inventory = []
        
    def add_book(self, book: Book) -> None:
        """Add new book to library inventory"""
        self._inventory.append(book)
        
    def checkout_book(self, title: str) -> Book:
        """
        Checkout first available copy of a book
        Returns checked out Book instance
        """
        for book in self._inventory:
            if book.title.lower() == title.lower() and book.status == "available":
                book.status = "borrowed"
                return book
        raise BookNotAvailableError(f"All copies of '{title}' are currently borrowed")
        
    def return_book(self, title: str) -> Book:
        """
        Return first borrowed copy of a book
        Returns returned Book instance
        """
        for book in self._inventory:
            if book.title.lower() == title.lower() and book.status == "borrowed":
                book.status = "available"
                return book
        raise BookNotFoundError(f"No borrowed copies of '{title}' found")
        
    def search_by_title(self, title: str) -> list[Book]:
        """Case-insensitive partial title search"""
        search_term = title.lower()
        return [book for book in self._inventory 
                if search_term in book.title.lower()]

# Usage Example
if __name__ == "__main__":
    # Initialize library
    lib = Library()
    
    # Add books
    lib.add_book(Book("1984", "George Orwell"))
    lib.add_book(Book("1984", "George Orwell"))  # Second copy
    lib.add_book(Book("Animal Farm", "George Orwell"))
    lib.add_book(Book("Brave New World", "Aldous Huxley"))

    # Test checkout system
    try:
        first_copy = lib.checkout_book("1984")
        print(f"Checked out: {first_copy}")
        
        second_copy = lib.checkout_book("1984")
        print(f"Checked out: {second_copy}")
        
        # Should fail - no available copies
        third_copy = lib.checkout_book("1984")
    except BookNotAvailableError as e:
        print(f"Error: {e}")

    # Test returns
    returned = lib.return_book("1984")
    print(f"Returned: {returned}")

    # Test search
    print("\nSearch results for 'world':")
    for book in lib.search_by_title("world"):
        print(f" - {book}")

    # Test invalid return
    try:
        lib.return_book("Non-existent Book")
    except BookNotFoundError as e:
        print(f"\nReturn Error: {e}")
```

###6. Property Decorators
Implement Temperature class:

Private attribute _celsius

Property celsius with getter/setter

Property fahrenheit with conversion logic

Validate values (-273.15°C minimum)

Target Concept: Properties, validation

```python
class Temperature:
    def __init__(self, celsius: float):
        self.celsius = celsius  # Uses the setter for validation

    @property
    def celsius(self) -> float:
        """Get temperature in Celsius"""
        return self._celsius

    @celsius.setter
    def celsius(self, value: float) -> None:
        """Set temperature in Celsius with validation"""
        if value < -273.15:
            raise ValueError("Temperature below absolute zero (-273.15°C) is impossible")
        self._celsius = value

    @property
    def fahrenheit(self) -> float:
        """Convert Celsius to Fahrenheit"""
        return (self._celsius * 9/5) + 32

    @fahrenheit.setter
    def fahrenheit(self, value: float) -> None:
        """Set temperature using Fahrenheit value"""
        self.celsius = (value - 32) * 5/9  # Reuse celsius setter for validation

    def __repr__(self) -> str:
        return f"Temperature({self._celsius:.1f}°C / {self.fahrenheit:.1f}°F)"

# Example usage
if __name__ == "__main__":
    try:
        # Valid temperature
        temp = Temperature(0)
        print(f"Initial: {temp}")  # 0°C / 32°F
        
        # Set via Celsius
        temp.celsius = 25
        print(f"After Celsius change: {temp}")  # 25°C / 77°F
        
        # Set via Fahrenheit
        temp.fahrenheit = 212
        print(f"After Fahrenheit change: {temp}")  # 100°C / 212°F
        
        # Invalid temperature
        temp.celsius = -300  # Triggers ValueError
        
    except ValueError as e:
        print(f"Error: {e}")
```
