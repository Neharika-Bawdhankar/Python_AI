# 2. List Element Sum
Problem: Create a function to return the sum of all even numbers in a list.
Example:
Input: [1][2][3][4][5]
Output: 6

```python
def even_numbers(list1):
    total = 0
    for element in list1:
        if element % 2 == 0:
            total = element + total
    return total

list1 = []
# list1 = [1,2,3,4,5]
Area_Circle = even_numbers(list1)
print(Area_Circle)
```

#4. Merge Lists Alternately
Problem: Combine two lists by alternating elements. If lists are unequal, append remaining elements.
Example:
Input: [1][2][3], ["a", "b"]
Output: [1, "a", 2, "b", 3]

```python
def merging_lists(seq1,seq2):
    mergered_list = []
    max_len = max(len(seq1), len(seq2))
    for i in range(max_len+1):
        if i < len(seq1):
            mergered_list.append(seq1[i])
        if i < len(seq2):
            mergered_list.append(seq2[i])
    return mergered_list
            
seq1 = [1,2,3]
seq2 = ['a','b']
Area_Circle = merging_lists(seq1, seq2)
print(Area_Circle)
```

##1. Arithmetic Operations Function
Problem:
Create a function perform_operation(a, b) that:

Adds a and b if they are integers/floats

Concatenates them if they are strings

Multiplies if a is a string and b is an integer (e.g., "Hi" * 3 → "HiHiHi")

```python
def perform_operation(a,b):
    if isinstance(a,(int,float)) and isinstance(b,(int,float)):
        return a + b
    if isinstance(a,str) and isinstance(b,str):
        return a + b
    if isinstance(a,str) and isinstance(b,int):
        return a * b
        
    
    
a = 'a'
b = 'b'   
Area_Circle = perform_operation(a,b)
print(Area_Circle)
```

#3. Animal Sound Generator

Create classes Dog, Cat, and Cow, each with a sound() method. Write a function make_sound(animal) that calls sound() on the input object.

```python
class Dog:
    
    def sound(self):
        return 'Woof!!'
                
class Cat:

    def sound(self):
        return 'Meaow!!'
                
class Cow:
    
    def sound(self):
        return 'Moo!!!'
     
def makes_sound(animal):
    return animal.sound()

dog = Dog()
print(makes_sound(dog))

cat = Cat()
print(makes_sound(cat))

cow = Cow()
print(makes_sound(cow))
```
