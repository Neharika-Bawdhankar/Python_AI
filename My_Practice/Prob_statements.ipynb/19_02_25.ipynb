# Problem: Given two integers, return their product if it's ≤1000. Otherwise, return their sum.


```python
def check_value(x,y):
    product = x * y
    sum = x + y
    if (product <= 1000):
        print("Product of X and Y is: ", product)
    else:
        print("Sum of X and Y is: ", sum)
        
check_value(30,40)
check_value(40,20)
```

```python
# Optimised version

def check_value(x, y):
    result = x * y if x * y <= 1000 else x + y
    print(f"{'Product' if x * y <= 1000 else 'Sum'} of {x} and {y} is: {result}")

check_value(30, 40)
check_value(40, 20)
```

# 2. String Reversal
Problem: Reverse a string without using [::-1] or reversed().
Input: "hello" → Output: "olleh"

```python
def reverse_string(str):
    count = len(str)-1
    str1 = []
    while count>=0:
        str1.append(str[count])
        count -= 1
    print(str1)
        
reverse_string("hello")
```

```python
# optimised version

def reverse_string(s):
    return ''.join([s[i] for i in range(len(s)-1, -1, -1)])

print(reverse_string("hello"))  # Output: 'olleh'
```

# 3. List Deduplication
Problem: Remove duplicates from a list without using set() or libraries.
Input: [1][2][3][2][1] → Output: [1][2][3]

```python
def check_duplicate(list1):
    li1 = []
    for i in list1:
        if i not in li1:
            li1.append(i)
        else:
            continue    
    print(li1)
    
    
list1 = [1,2,3,2,1]
check_duplicate(list1)   
```

```python
# optimised version


def remove_duplicates(input_list):
    seen = {}
    return [x for x in input_list if not (x in seen or seen.update({x: True}))]
```

#4. Palindrome Check
Problem: Determine if a string is the same forwards and backwards (case-sensitive).
Input: "Racecar" → Output: False
Input: "madam" → Output: True

```python
def palindrome(str):
    reverse_string = ''.join(str[i] for i in range(len(str)-1,-1,-1))
    # print(reverse_string)
    if reverse_string == str:
        return True
    return False
    
str = 'Racecar'
result = palindrome(str)
print(result)



# is_palindrome = all(s[i] == s[~i] for i in range(len(s)//2))
```

#6. Vowel Counter
Problem: Count vowels (a, e, i, o, u) in a string.
Input: "Hello World" → Output: 3


```python
def count_vowels(word):
    count = 0
    vowels = 'aeiou'
    for alphabet in word:
        if alphabet in vowels:
            count += 1
        # else:
        #     continue
    return count

word = 'Hello World'
result = count_vowels(word)
print(result)
```

#8. Digit Sum Calculator
Problem: Calculate the sum of all digits in a number.
Input: 1234 → Output: 10 (1+2+3+4)

```python
def sum_calculator(numbers):
    sum = 0
    for i in numbers:
        sum = sum + i
    return sum
    


numbers = 1,2,3,4,1
sum_calculator(numbers)
```

```python
def is_prime(number):
    if number <= 1:
        return False
    
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

# Test cases
print(is_prime(7))   # Output: True
print(is_prime(9))   # Output: False
print(is_prime(2))   # Output: True
print(is_prime(4))   # Output: False
print(is_prime(25))  # Output: False
print(is_prime(97))  # Output: True
```
