```python
# Shallow copy of List using list()

a = [1, 2, 3]
a1 = list(a)

a1[0] = 5
print(a)   # Output: [1, 2, 3]

# different case for Nested List
b = [[1, 2, 3], [4, 5, 6]]
b1 = list(b)

# changing b1 will also reflect in b
b1[0][0] = 10
print(b)  # Output: [[10, 2, 3], [4, 5, 6]]
```

```python
#Creating Nested List with list() constructor

mat = list([list(range(3)) for _ in range(3)]) 
print(mat)
```

```python
#Using else statement with While Loop in Python
cnt = 0
while (cnt < 3):
    cnt = cnt + 1
    print("Hello Geek")
else:
    print("In Else Block")
    
    
```

```python
#Infinite While Loop in Python
count = 0
while (count == 0):
    print("Hello Geek")

# Note: It is suggested not to use this type of loop as it is a never-ending infinite loop where the condition is always true and we have to forcefully terminate the compiler.
```
