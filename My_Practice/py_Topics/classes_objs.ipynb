```python
#Build in functions - polymorphism
def add(a,b):
    return a + b

print(add(3,4))
print(add('a', 'b'))
```

```python
def add(a, b):
    return a + b

print(add(3, 4))           # Integer addition
print(add("Hello, ", "World!"))  # String concatenation
print(add([1, 2], [3, 4])) # List concatenation
```

```python
class Dog:
    def __init__(self, name, breed, age):
        self.name = name 
        self._breed = breed
        self.__age = age
    # public method
    def get_info(self):
        return f"Name : {self.name}, Breed: {self._breed}, Age: {self.__age}"

    def get_age(self):
        return self.__age
    
    def set_age(self, age):
        if age > 0:
            self.__age =  age
        else:
            print("Invalid age!")
            
dog1 = Dog("Buddy","Labrador",3)
print(dog1.name)
print(dog1._breed)

        
```
