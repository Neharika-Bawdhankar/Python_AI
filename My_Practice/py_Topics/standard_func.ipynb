```python
# An integer
var = -94
print('Absolute value of integer is:', abs(var))
```

```python
# floating point number
float_number = -54.26
print('Absolute value of float is:', 
                  abs(float_number))
```

```python
# A complex number
complex_number = (3 - 4j)
print('Absolute value or Magnitude of complex is:', abs(complex_number))
```

```python
# All elements of list are true
l = [4, 5, 1]
print(all(l))

# All elements of list are false
l = [0, 0, 0]
print(all(l))

# Some elements of list are
# true while others are false
l = [1, 0, 6, 7, False]
print(all(l))

# Empty List
l = []
print(all(l))

# all() with condition - to check if all elements are greater than 0
l = [1,-3,0,2,4]
print(all(ele > 0 for ele in l))
```

```python
# Python program to illustrate 
# callable() a test function
def Geek():
	return 5

# an object is created of Geek()
let = Geek
print(callable(let))

# a test variable
num = 5 * 5
print(callable(num))
```
