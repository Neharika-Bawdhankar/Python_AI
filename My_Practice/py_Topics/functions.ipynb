```python
#Polymorphism functions

def multiply_nums(num1,num2):
    return num1 * num2
    
# multiply_nums(2,10)
# multiply_nums('a',5)
multiply_nums(4,'A')
```

```python
#args    # type = tuple

def demo(*args):
    return sum(args)

demo(1,23,4,5,7,8)
    
#  **kwargs    # type = dict

def demo(**kwargs):
    for key,value in kwargs.items():
        print(key, value)
    for key in kwargs.keys():
        print(key)
    for value in kwargs.values():
        print(value)
       
demo(a=1, b=2, c=3)

```

```python
def my_function(*args):
    for arg in args:
        print(arg)
        
my_function('Hello', 'Welcome', 'to', 'GeeksforGeeks','USA')
```

```python
def fun(arg1, *argv):
    print("First argument :", arg1)
    for arg in argv:
        print("Argument *argv :", arg)


fun('Hello', 'Welcome', 'to', 'GeeksforGeeks')
```

```python
#using both args and kwargs

def fun(*args, **kwargs):
    print("positional arguments:", args)
    print("keyword arguments:" ,  kwargs)
    
fun(1,2,2,a=1,b=2)
```
