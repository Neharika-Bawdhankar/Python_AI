```python
class Person(BaseModel):
    name: str
    age: int
    email: str

person=Person(name="Neharika",age=35,email="abc@gmail.com")
print(person)
```

```python
from typing import Optional

class Employee(BaseModel):
    name: str
    department: str
    salary: Optional[float] = None
    is_active: Optional[bool] = True
    
person = Employee(name="Keval", department="IT", salary=50000000)
print(person)
```

```python
from pydantic import BaseModel, Field
class Product(BaseModel):
    name: str=Field(min_length=3, max_length=50)
    price: float=Field(gt=0, description="Price must be greater than zero")
    quantity: int=Field(ge=0, description="Quantity must be greater than or equal to zero")
    
    
item = Product(name="Laptop", price=1000.0, quantity=7)
print(item)
```

```python
from pydantic import BaseModel, Field
class Product(BaseModel):
    name: str=Field(min_length=3, max_length=50)
    price: float=Field(gt=0, description="Price must be greater than zero")
    quantity: int=Field(ge=0, description="Quantity must be greater than or equal to zero")
    
    
item = Product(name="Laptop", price=1000.0, quantity=-7)
print(item)


    
```
